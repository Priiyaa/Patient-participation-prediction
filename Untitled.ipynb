```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df=pd.read_csv("Patient participation file2.csv")
```

```python
df.head()
```

```python
table=pd.crosstab(df.Mobility,df.Participation)
table.div(table.sum(1).astype(float), axis=0).plot(kind='bar', stacked=True)
plt.title('Mobiity vs age')
plt.xlabel('Mobility')
plt.ylabel('Participation')
plt.savefig('edu_vs_pur_stack')
```

```python
from sklearn.preprocessing import LabelEncoder
le_mobility = LabelEncoder()
df['Mobility']=le_mobility.fit_transform(df['Mobility'])
df["Mobility"].unique()
```

```python
plt.title('Age vs Participation')
plt.xlabel('Age')
plt.ylabel('Participation')
plt.scatter(df.Participation,df.Age)
plt.show()
plt.title('Age vs Mobility')
plt.xlabel('Age')
plt.ylabel('Mobility')
plt.scatter(df.Age,df.Mobility)
plt.show()
```

```python
df_parameters = pd.DataFrame(dict(
    Mobility=df['Mobility'], 
    Age=df['Age'],
    Distance=df['Distance']
))
df_parameters
```

```python
X=df_parameters
print(X.ndim)

y=df['Participation']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report

# Create training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=0)

# Create the classifier: logreg
logreg = LogisticRegression()

# Fit the classifier to the training data
logreg.fit(X_train, y_train)

# Predict the labels of the test set: y_pred
y_pred = logreg.predict(X_test)

# Compute and print the confusion matrix and classification report
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
